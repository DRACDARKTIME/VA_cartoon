```python
cartoon_pngcsv = path.data_raw_dir('cartoonset10k/cartoonset10k')
cartoon_png = path.data_processed_dir('cartoon_png')
if cartoon_png.exists()==False:
    os.makedirs(cartoon_png)
```

```python
for file_name in os.listdir(cartoon_pngcsv):
    if file_name.endswith('.png'):
        source = cartoon_pngcsv.joinpath(file_name)
        destination= cartoon_png.joinpath(file_name)
        os.rename(source, destination)
```
